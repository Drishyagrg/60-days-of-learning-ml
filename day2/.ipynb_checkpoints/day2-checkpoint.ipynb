```python
import pandas as pd
import requests
```

# fetching data from an api

```python
url = "https://anime-db.p.rapidapi.com/anime"

querystring = {"page":"1","size":"50","sortOrder":"asc"}

headers = {
	"x-rapidapi-key": "rapid_api_key",
	"x-rapidapi-host": "anime-db.p.rapidapi.com",
	"Content-Type": "application/json"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
temp_df = pd.DataFrame(response.json()['data'])[['id','title','alternativeTitles','genres','ranking','synopsis','episodes','status','type']]
```

```python
all_frames =[]
for i in range(1,602):
    url = "https://anime-db.p.rapidapi.com/anime"
    querystring = {"page":'{}'.format(i),"size":"50","sortOrder":"asc"}
    headers = {
    	"x-rapidapi-key": "rapid_api_key",
    	"x-rapidapi-host": "anime-db.p.rapidapi.com",
    	"Content-Type": "application/json"
    }
    response = requests.get(url, headers=headers, params=querystring)
    
    temp_df = pd.DataFrame(response.json()['data'])[['id','title','alternativeTitles','genres','ranking','synopsis','episodes','status','type']]
    all_frames.append(temp_df)
    
df = pd.concat(all_frames, ignore_index=True)
    
```

```python
df
```

```python
df.to_csv("anime.csv")
```
