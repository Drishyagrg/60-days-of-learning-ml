```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# EDA using Bivariate and Multivariate

```python
df = pd.read_csv("anime_dataset.csv")
```

## 1) Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(y= df['episodes'], x=df['score'],hue=df['rank'])
```

## 2) Bar plot (Numerical - Categorical)

```python
sns.barplot(x=df['type'],y=df['episodes'],hue=df['airing'])
```

## 3) Box plot (numerical - categorical)

```python
sns.boxplot(x=df['type'], y=df['popularity'])
```

## 4) Distplot (numerical - categorical)

```python
sns.distplot(df['score'])
```

## 5) Heatmap (categorical - categorical)

```python
sns.heatmap(pd.crosstab(df['type'],df['source']))
```

## 6) Clustermap (categorical - categorical)

```python
sns.clustermap(pd.crosstab(df['source'], df['season']))
```

## 7) Pairplot (we can say collection of scatterplot)
sns.pairplot(df)

## 8) Lineplot (numerical - numerical )
x-axis -> time based

```python
sns.lineplot(x=df['year'], y=df['members'])
```
